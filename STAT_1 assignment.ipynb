```python
import statistics 
wx = [0, 1, 2, 3, 4, 5]
wfx =[ 0.09, 0.15, 0.40, 0.25, 0.10, 0.01]
y=[a*b for a,b in zip(wx,wfx)]
mean=nm.sum(y)
#l=[(i**2)*j for i,j in zip(wx,wfx)]
#var=nm.sum(l)-mean**2
z=[((i-mean)**2)*j for i,j in zip(wx,wfx)]
d=nm.sum(z)
print('mean = {0}'.format(mean),'variance = {0}' .format(d),sep='\n')
```
